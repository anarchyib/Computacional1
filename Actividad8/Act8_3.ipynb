```python
from numpy import *
import pylab as p
# Definition of parameters
mu= 5.0
def dX_dt(X, t=0):
    xp=X[1]
    yp=mu*(1-X[0]**2)*X[1]-X[0]
    return array([xp, yp ])
```

```python
from scipy import integrate
t = linspace(10, 50,  4000)              # time
X0 = array([2, 0])                     # initials conditions: 10 rabbits and 5 foxes
X, infodict = integrate.odeint(dX_dt, X0, t, full_output=True)
infodict['message']
```

```python
from numpy import loadtxt
from pylab import figure, plot, xlabel, grid, hold, legend, title, savefig, xlim, ylabel, ylim
from matplotlib.font_manager import FontProperties
%matplotlib inline
x, y = X.T

plot(t, x, 'c-')
grid()
xlim(10,50)
ylim(-3,3)
xlabel('tiempo')
ylabel('x')
title('Posición contra tiempo')
savefig('Posicion_contra_tiempo.png')
```
