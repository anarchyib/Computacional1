```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
sal =pd.read_csv('sargento-salinidad-201117.csv',skiprows=4,sep=',', header=None, names=['#', 'DATETIME', 'CHR', 'TEMP', 'SPECOND', 'SALINITY'])
```

```python
sal['Ndate'] = pd.to_datetime(sal['DATETIME'], format='%m/%d/%Y %H:%M:%S')
sal['month'] = sal['Ndate'].dt.month
sal.head()
```

```python
df1 =pd.read_csv('sargento_201117.csv',skiprows=3,sep=',', header=None, names=['#', 'DATETIME', 'PRES', 'TEMP', 'WATERLEVEL'])
```

```python
df1.head()
```

```python
df1['Ndate'] = pd.to_datetime(df1['DATETIME'], format='%m/%d/%Y %H:%M:%S')
df1['month'] = df1['Ndate'].dt.month
df1.head()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import rc
rc('mathtext', default='regular')

fig = plt.figure()
ax = fig.add_subplot(111)

ax.plot(df1["#"], df1["WATERLEVEL"], '-', label = 'Nivel')
ax2 = ax.twinx()
ax2.plot(df1["#"], df1["TEMP"], '-r', label = 'Temperatura')

# ask matplotlib for the plotted objects and their labels
lines, labels = ax.get_legend_handles_labels()
lines2, labels2 = ax2.get_legend_handles_labels()
ax2.legend(lines + lines2, labels + labels2, loc=0)

ax.grid()
ax.set_xlabel("Fecha")
ax.set_ylabel(r"Nivel del Agua (metros)")
ax2.set_ylabel(r"Temperatura (C)")
ax2.set_ylim(0, 25)
ax.set_ylim(-10,25)
plt.show()
```
